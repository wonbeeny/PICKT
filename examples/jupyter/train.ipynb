```python
import os
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

from pickt import Worker, load_data
```

```python
os.chdir("../../..")
config = load_data("./PICKT/examples/config/pickt_train_config.yaml")

km_data = load_data(config.km_data_path)
data_args = load_data(config.data_args_path)
train_datasets = load_data(config.train_dataset_path)
valid_datasets = load_data(config.valid_dataset_path)

train_datasets["data_args"] = data_args
train_datasets["gnn_data"] = km_data
valid_datasets["gnn_data"] = km_data
```

```python
worker = Worker(
    config, 
    train_datasets=train_datasets, 
    valid_datasets=valid_datasets
)

worker.train()
```
